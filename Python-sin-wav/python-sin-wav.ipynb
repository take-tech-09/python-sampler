# Pythonで正弦波(sin波)を作成
https://take-tech-engineer.com/python-sin-wav/

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
A = 0.5    # 振幅
f = 1.0    # 周波数 Hz
sec = 3.0  # 信号の長さ s
sf = 44100 # サンプリング周波数 Hz
```

```python
t = np.arange(0, sec, 1/sf) #サンプリング点の生成

y = A*np.sin(2*np.pi*f*t) # 正弦波の生成
```

```python
plt.plot(t, y);
```

```python
t = torch.arange(0, sec, 1/sf) #サンプリング点の生成

y = A*torch.sin(2*np.pi*f*t) # 正弦波の生成
```

```python
plt.plot(t, y);
```
