```python
import numpy as np
import torch

import matplotlib.pyplot as plt
```

```python
A = 1.0    # 振幅
sec = 3.0  # 信号の長さ s
sf = 44100 # サンプリング周波数 Hz
```

```python
x = 2*A*(np.random.rand(round(sf*sec))-0.5) # ホワイトノイズの生成

plt.plot(x); # 波形の表示
```

```python
x = 2*A*(torch.rand(round(sf*sec))-0.5) # ホワイトノイズの生成

plt.plot(x); # 波形の表示
```
